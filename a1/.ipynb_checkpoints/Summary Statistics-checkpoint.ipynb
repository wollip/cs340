```julia
import Statistics
import StatsBase
```

```julia
using DelimitedFiles
dataTable = readdlm("fluTrends.csv", ',');
```

# Summary statistics

```julia
num_data, num_features = size(dataTable)
num_data -= 1
for i in 1:num_features
    println(dataTable[1, i])
    println("\tmin:\t", minimum(dataTable[2:end, i]))
    println("\tmax:\t", maximum(dataTable[2:end, i]))
    println("\tmean:\t", Statistics.mean(dataTable[2:end, i]))
    println("\tmode:\t", StatsBase.mode(dataTable[2:end, i]))
end
```
