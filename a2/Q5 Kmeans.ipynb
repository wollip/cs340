```julia
# Load data
using JLD
X = load("clusterData.jld","X");
include("kMeans.jl");
```

```julia
# K-means clustering
errors = zeros(10)
for k in 1:10
    min_e = 1000000000
    for i in 1:50
        model = kMeans(X,k,doPlot=false)
        error = kMeansError2(X, model.y, model.W)
        min_e = min(min_e, error)
    end
    errors[k] = min_e
end
```

```julia
plot(1:10, errors)
xlabel("K")
ylabel("Error")
```
